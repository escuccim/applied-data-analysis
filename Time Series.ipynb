```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn
seaborn.set()
```

```python
diff=data-data.shift(1)
diff.plot(figsize=(16,12))
```

```python
roi = 100 * (data.shift(-365)/data-1)
roi.plot(figsize=(16,12))
```

```python
data.plot(figsize=(16,12), style='-')
data.resample('BA').mean().plot(style=':')
data.asfreq('BA').plot(style='--')
plt.legend(['input', 'resample', 'asfreq'], loc='upper left');
```

```python
data.asfreq('D', method='bfill')[0:100].plot(figsize=(16,12), style='-o')
data.asfreq('D', method='ffill')[0:100].plot(style='--o')
plt.legend(['bfill', 'ffill'],
           loc='upper left');
plt.show()
```

```python
rolling = data.rolling(365, center=True)
data.plot(figsize=(16,12))
rolling.mean().plot(color='red', linewidth=3)
```

```python
data_log = data.apply(np.log)
```

```python
data_log.head()
```

```python
variance1 = data.rolling(30, center=True).var() 
variance2 = data_log.rolling(30, center=True).var() 
```

```python
fig, ax = plt.subplots(2,1, figsize=(13,12))
variance1.plot(ax=ax[0], title='Original Variance')
variance2.plot(ax=ax[1], title='Log Variance')
```

```python
import pandas as pd
data = pd.read_csv('weatherHistory.csv', index_col='Date', parse_dates=True)
data.head()
```

```python
data['datetime'] = data.index
```

```python
data.datetime.mean()
```

```python
data["Hour"] = data.index.hour
```

```python
data.head()
```

```python
hour_avg = data.groupby("Hour")["Temperature (C)"].mean()
```

```python
data["Month"] = data.index.month
```

```python
mn_avg = data.groupby("Month")["Temperature (C)"].mean()
```

```python
mn_avg
```

```python
fig, ax = plt.subplots(1,2, figsize=(13,12))
hour_avg.plot(ax=ax[0], title='Hourly Mean Temp')
mn_avg.plot(ax=ax[1], title='Monthly Mean Temp')
plt.show()
```
